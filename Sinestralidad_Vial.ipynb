```python
import pandas as pd
df = pd.read_csv("/content/SECTORES_CRITICOS_DE_SINIESTRALIDAD_VIAL_20241110.csv")

# Total de muertes
total_muertes = df['Fallecidos'].sum()
print(f"Total de muertes: {total_muertes}")


# Tramo con el mayor GiPValue
tramo_max_gipvalue = df.loc[df['GiPValue'].idxmax(), 'Tramo']
max_gipvalue = df['GiPValue'].max()
print(f"Tramo con el mayor GiPValue: {tramo_max_gipvalue} con un valor de {max_gipvalue}")

# Tramo con el mayor GiZScore
tramo_max_gizscore_idx = df['GiZScore'].idxmax()
tramo_max_gizscore = df.loc[tramo_max_gizscore_idx, 'Tramo']
max_gizscore = df['GiZScore'].max()

# Total de fallecidos en ese tramo (sumando los fallecidos de todas las filas con ese tramo)
total_fallecidos_tramo_max_gizscore = df[df['Tramo'] == tramo_max_gizscore]['Fallecidos'].sum()

print(f"Tramo con el mayor GiZScore: {tramo_max_gizscore} con un valor de {max_gizscore}")
print(f"Total de fallecidos en ese tramo: {total_fallecidos_tramo_max_gizscore}")

# Agrupar por tramo y sumar los fallecidos
fallecidos_por_tramo = df.groupby('Tramo')['Fallecidos'].sum()

# Encontrar el tramo con el mayor número de fallecimientos
tramo_max_fallecidos = fallecidos_por_tramo.idxmax()
max_fallecidos = fallecidos_por_tramo.max()

# Obtener el valor de PR para ese tramo
pr_tramo_max_fallecidos = df[df['Tramo'] == tramo_max_fallecidos]['PR'].iloc[0]

print(f"Tramo con el mayor número de fallecimientos: {tramo_max_fallecidos} con {max_fallecidos} fallecidos")
print(f"Valor de PR para ese tramo: {pr_tramo_max_fallecidos}")

# Agrupar por Latitud y Longitud y sumar los Fallecidos
muertes_por_ubicacion = df.groupby(['Latitud', 'Longitud'])['Fallecidos'].sum()

# Encontrar la ubicación con más fallecimientos
ubicacion_max_fallecidos = muertes_por_ubicacion.idxmax()
max_fallecidos_ubicacion = muertes_por_ubicacion.max()

print(f"Ubicación con más muertes: Latitud = {ubicacion_max_fallecidos[0]}, Longitud = {ubicacion_max_fallecidos[1]}")
print(f"Total de muertes en esa ubicación: {max_fallecidos_ubicacion}")

# Agrupar por Departamento y sumar los Fallecidos
muertes_por_departamento = df.groupby('Departamento')['Fallecidos'].sum()

# Encontrar el departamento con más muertes
departamento_max_fallecidos = muertes_por_departamento.idxmax()
max_fallecidos_departamento = muertes_por_departamento.max()

# Calcular el promedio de GiZScore para ese departamento
promedio_gizscore_departamento = df[df['Departamento'] == departamento_max_fallecidos]['GiZScore'].mean()

print(f"Departamento con más muertes: {departamento_max_fallecidos} con {max_fallecidos_departamento} fallecidos")
print(f"Promedio de GiZScore para ese departamento: {promedio_gizscore_departamento}")

# Filtrar los valores válidos de GiPValue (excluyendo NaN y ceros)
giPValue_validos = df['GiPValue'][(df['GiPValue'] != 0) & (df['GiPValue'].notna())]

# Calcular el promedio de GiPValue
promedio_gipvalue = giPValue_validos.mean()

print(f"Promedio de GiPValue (sin NaN ni ceros): {promedio_gipvalue}")

# Agrupar por ENTIDAD y sumar los Fallecidos
muertes_por_entidad = df.groupby('ENTIDAD')['Fallecidos'].sum()

# Mostrar el resultado
print("Número de muertes por entidad:")
print(muertes_por_entidad)

# Agrupar por Tramo y sumar los Fallecidos
muertes_por_tramo = df.groupby('Tramo')['Fallecidos'].sum()

# Ordenar por el número de muertes de mayor a menor y obtener los 5 tramos con más muertes
top_5_tramos = muertes_por_tramo.sort_values(ascending=False).head(5)

# Mostrar el resultado
print("Los 5 tramos con más muertes:")
print(top_5_tramos)

```
